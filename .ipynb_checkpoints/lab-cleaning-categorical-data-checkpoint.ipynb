```python
#1. Import the necessary libraries.
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
customer_df = pd.read_csv('files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv')
customer_df.head()
```

```python
#2. Find all of the categorical data. Save it in a categorical_df variable.
categorical_df=customer_df.select_dtypes(object)
```

```python
#3. Check for NaN values.
categorical_df.isna().sum()/len(categorical_df)
```

```python
# 4. Check all unique values of columns.
# (We effectively excluded Customer and Effective To Date columns as they were ID column and Date column respectively.)
for col in categorical_df.columns:
    if col!='Customer' and col!='Effective To Date':
        print(categorical_df[col].value_counts())
        print()
```

```python
#5. Check dtypes. Do they all make sense as categorical data?
```

Excluding Customer and Effective to Date, the values in other columns are distinct and countable as expected from a categorical column and none of the column values are specified as integer within the categorical column. 

The Customer column is an ID column containing alpha numeric and Effective to Date is effectively a Date column and needs to converted to one.

Other categorical column have scope for revision and transformation to other kinds (like ordinal etc.)

```python
categorical_df['Effective To Date'] =  pd.to_datetime(categorical_df['Effective To Date'])
categorical_df.head()
```

```python
#6. Does any column contain alpha and numeric data? Decide how to clean it.
```

The column policy and offer contains alpha and numeric data. On observation, it is found that the numeric values in these columns are nothing but the levels, and string values seem redundant. The numerical values can be converted to ordinal values through transformation. (The column policy will be converted later in #8.)

```python
categorical_df['Renew Offer Type']=categorical_df['Renew Offer Type'].apply(lambda x: x[-1])
categorical_df.head()
```

```python
#7. Would you choose to do anything else to clean or wrangle the categorical data? Comment your decisions.
```

State - They are countable to 5 unique values, so we wouldn't do anything to this column, until we start observing that more state values start pooling in this column and count of those state values remain less than 1000, we can effectively group them to a value called Others.

Policy and Offer - As explained in the above question, they will need to be converted to ordinal values.

Vehicle Class - The column values can be combined to a fewer distinct values by grouping similar class/segment of vehicles.


```python
#8. Compare policy_type and policy. What information is contained in these columns. Can you identify what is important?
```

The column policy contains redundant data and the similar information is already available in the column policy type. The two columns are tighly correlated to each other. That is, the only value expected in the policy column is the policy level. The column name policy can later be renamed as policy level.

Before making the change to column Policy, we make sure the 2 column values are indeed related to each other (although it's obvious). That is we make sure that the common value in both the columns (values between Personal, Corporate, Special) is equal.

```python
if (categorical_df['Policy Type'].apply(lambda x: x.split(" ")[0])==categorical_df['Policy'].apply(lambda x: x.split(" ")[0])).all():
    categorical_df['Policy']=categorical_df['Policy'].apply(lambda x: x[-1])

categorical_df.rename(columns={'Policy': 'Policy Level'}, inplace=True)
categorical_df.head()
```

```python
def standardize_vehicle_class(x):
    if x in ['Two-Door Car', 'Four-Door Car']:
        return 'Standard'
    elif x in ['SUV', 'Sports Car']:
        return 'Sports'
    else:
        return 'Luxury'
    
categorical_df['Vehicle Class']=categorical_df['Vehicle Class'].apply(standardize_vehicle_class)
categorical_df['Vehicle Class'].value_counts()
```
